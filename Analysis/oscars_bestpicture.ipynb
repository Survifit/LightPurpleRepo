```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
primary_csv = "../Data/clean_data.csv"
```

```python
#Reading data file 
primary_df = pd.read_csv(primary_csv)
```

```python
oscars_csv = "../Data/oscars.csv"
```

```python
oscars_df = pd.read_csv(oscars_csv)
```

```python
oscars_df.head()
```

```python
primary_df['bestpicture'] = primary_df['yesno']
primary_df
```

```python
primary_df
```

```python
oscars_df['oscartitle'] = oscars_df.oscartitle.str.replace(r'"', '')
```

```python
oscarlist = oscars_df['oscartitle'].tolist()
```

```python
oscarlist
```

```python


primary_df['bestpicture'] = primary_df.Title.isin(oscarlist)


        


```

```python
oscars_df = primary_df.loc[primary_df['bestpicture'] == True]
oscars_df
```

```python
oscars_df.mean()
```

```python
non_oscars_df = primary_df.loc[primary_df['bestpicture'] == False]
```

```python
non_oscars_df.mean()
```

```python
#T-test indicates that we should reject null hypothesis of no difference in the means of the IMDB Rating between the two groups (movies scoring zero on bechdel and movies scoring three on bechdel)
stats.ttest_ind(oscars_df.rating, non_oscars_df.rating, equal_var=False)
```
